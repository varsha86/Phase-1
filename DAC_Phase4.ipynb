```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
product=pd.read_csv("statsfinal.csv")
product
```

```python
mean = np.mean(product)
mean
```

```python
std_dev = np.std(product)
std_dev
```

```python
from matplotlib import pyplot as plt
x = mean
y = std_dev
plt.plot(mean,std_dev)
plt.title("Line graph")
plt.ylabel('Y axis')
plt.xlabel('X axis')
plt.show()
```
